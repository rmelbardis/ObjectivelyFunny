```python
import pandas as pd
```

```python
df = pd.read_json('raw_data/live_at_the_apollo.json')
```

```python
df = df.reset_index()
```

```python
df
```

```python
df['host_script'] = df['full_transcript']
#[x.split(df['Comedian1'].string)[-0] for x in df['full_transcript']]
df['com1_script'] = df['full_transcript']
#[x.split(df['Comedian1'].string)[-1] for x in df['full_transcript']]
df['com2_script'] = df['full_transcript']
#[x.split(df['Comedian2'].string)[-1] for x in df['full_transcript']]
```

```python
for i,row in df.iterrows():
    df['host_script'][i]= df['full_transcript'][i].split(str(df['Comedian1'][i]))[-0]
```

```python
df['host_script'][2]
```

```python
df['com1_script'] = df.apply(lambda row: row['full_transcript'].replace(str(row['host_script']),''),axis=1)
df['com2_script'] = df.apply(lambda row: row['full_transcript'].replace(str(row['host_script']),''),axis=1)
```

```python
for i,row in df.iterrows():
    df['com1_script'][i]= df['com1_script'][i].split(str(df['Comedian2'][i]))[-0]
```

```python
df['com2_script'] = df.apply(lambda row: row['com2_script'].replace(str(row['com1_script']),''),axis=1)
```

```python
df['com1_script'] = df.apply(lambda row: row['com1_script'].replace(str(row['patter']),''),axis=1)
```

```python
df['patter'] = df['com2_script']
for i,row in df.iterrows():
    df['patter'][i]= df['com2_script'][i].split(str(df['Comedian2'][i]))[-1]
```

```python
df['com2_script'] = df.apply(lambda row: row['com2_script'].replace(str(row['patter']),''),axis=1)
```
