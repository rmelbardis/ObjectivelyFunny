```python
import pandas as pd
```

```python
df = pd.read_json('../raw_data/temp_df_for_testing.json')
df.head()
```

```python
age_df = pd.DataFrame((df.age_then//10)*10)
age_df
```

```python
age_df['age_group'] = age_df['age_then'] 
age_df
```

```python
age_df['age_group'][age_df['age_group'].isnull()]=-1
```

```python
age_df['age_group'] = age_df['age_group'].astype(int)
```

```python
import numpy as np
```

```python
age_df['age_group'] [age_df['age_then']<0]= np.nan
age_df['age_group'] [age_df['age_then']<30]= 'U30'
age_df['age_group'] [age_df['age_then']>=60]= 'O60'
age_df
```

```python
age_df['age_group'] = age_df['age_group'].astype(str)
```

```python
age_df['age_group']
```

```python
age_group_df = pd.DataFrame(age_df['age_group'])
age_group_df
```

```python
pd.DataFrame.to_json(age_group_df, 'age_group.json')
```
