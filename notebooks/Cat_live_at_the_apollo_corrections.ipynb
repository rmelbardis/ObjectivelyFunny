```python
df['title'] = 'Live at the Apollo Series ' + df['series'].astype(str) + ' Episode ' + df['episode'].astype(str)
```

```python
final_df = df[['title', 'Host', 'Comedian1', 'Comedian2', 'Year', 'host_script', 'com1_script', 'com2_script' ]]
```

```python
final_df
```

```python
host_df = final_df[['title', 'host_script', 'Host', 'Year']]
com1_df = final_df[['title', 'com1_script', 'Comedian1', 'Year']]
com2_df = final_df[['title', 'com2_script', 'Comedian2', 'Year']]
```

```python
host_df = host_df.rename(columns = {'host_script': 'full_transcript', 'Host': 'artist'})
com1_df = com1_df.rename(columns = {'com1_script': 'full_transcript', 'Comedian1': 'artist'})
com2_df = com2_df.rename(columns = {'com2_script': 'full_transcript', 'Comedian2': 'artist'})
```

```python
full_df = pd.concat([host_df, com1_df, com2_df])
```

```python
full_df = full_df.reset_index(drop=True)
```

```python
full_df = full_df.drop(columns=['index'])
```

```python
full_df = full_df.drop([85])
```

```python
full_df.to_json('live_at_the_apollo.json')
```
