```python
import pandas as pd
```

```python
import re
import numpy as np
pattern = r'([A-Za-z0-9\s\.]*?):([A-Za-z0-9\s\.]*)\((\d{4})\)'
```

```python
test_1 = titles.iloc[0][0]
test_1
```

```python
break_titles = re.search(pattern, test_1)
break_titles
break_titles.group(1)
```

```python

def extract_artist(text):
    pattern = r'([A-Za-z0-9\s\.]*?):'
    response = re.search(pattern, text)
    if response == None:
        return np.nan
    return response.group(1)
def extract_show_name(text):
    pattern = r':([A-Za-z0-9\s\.]*)\('
    response = re.search(pattern, text)
    if response == None:
        return np.nan
    return response.group(1).strip()
def extract_year(text):
    pattern = r'\((\d{4})\)'
    response = re.search(pattern, text)
    if response == None:
        return np.nan
    return response.group(1)
```

```python
extract_artist(test_1)
```

```python
extract_show_name(test_1)
```

```python
extract_year(test_1)
```

```python
df['artist'] = df.title.apply(extract_artist)
```

```python
df.head()
```

```python
df['show_name'] = df.title.apply(extract_show_name)
```

```python
df['year'] = df.title.apply(extract_year)
```

### filling in missing show names

```python
df2 = df[df.show_name.isnull()]
df2
```

```python
df2.index
```

```python
new_list = [i[0] for i in df2.title.str.split('(')]
show_list = []
for char in new_list:
    temp = char.split(': ')[1]
    show_list.append(temp.replace('–', '').replace(
        '[TRANSCRIPT]', '').replace('[FULL TRANSCRIPT]','').replace(
        '[Transcript]','').replace('Full Transcript','').replace(
        'Transcript', '').replace('Full transcript', '').strip())
show_list
```

```python
df2['show_name'] = pd.Series(show_list, index=df2.index)
```
