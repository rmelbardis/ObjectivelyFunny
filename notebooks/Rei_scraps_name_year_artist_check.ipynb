```python
def capitals(text):
    words = text.split(' ')
    cap_words = [word.capitalize() for word in words]
    return ' '.join(cap_words)
```

```python
df['artist'] = df['artist'].apply(capitals)
df['show_name'] = df['show_name'].apply(capitals)
```
